```python
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_example = scipy.io.loadmat('../Example_data/example_results_forcast/brain_score_results.mat')
```

```python
avg_result = np.mean(results_example["brain_score"], axis=0)
# This makes the forcast score
norm_avg_result = avg_result - avg_result[:,0][:,np.newaxis]
```

```python
# Plot
fig, axes = plt.subplots(4, 8, figsize=(15, 10))  # Create a 4x8 grid of subplots
fig.suptitle("Channel-wise Time Series")

time = np.arange(25)  # Assuming time points are 0 to 15

for i, ax in enumerate(axes.flat):
    if i < 32:
        ax.plot(time, norm_avg_result[i, :], label=f"CH {i+1}")

        ax.set_title(f"CH {i+1}")
        ax.set_xlabel("Time")
        ax.set_ylabel("Value")
        ax.grid(True)

    else:
        ax.axis("off")  # Hide unused subplots

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit title
plt.show()
```
